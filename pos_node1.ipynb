```python
def main():
    """Run test"""
    account = {'Address': 'eltneg', 'Weight': 50}

    blockchain = Blockchain(GENESIS_BLOCK, account)
    blockchain.generate_new_block(52)
    clients = [blockchain]
    blockchain.get_blocks_from_nodes()
    blockchain.pick_winner()
    blockchain.pos()


    while True:
        print('============================================ \n\n')
        client = clients[randint(0, 0)]
        client.pos()
        time.sleep(3)
```
